### Votre implémentation

```
# Partie A : Génération des données
seed(71)                   # graine aléatoire
n = 1000                   # points sur l'axe horizontal

x = rand(n) * 10           # 1000 valeurs uniformes dans l'intervalle [0,10)
y = zeros(n)               # un vecteurs contenant 1000 zéros
points = array([x,y])      # matrice (2x1000) qui fusionnent x et y

# Matrice de rotation
teta = pi/6                                    # 30 dégrés
R_teta = array([[cos(teta), -sin(teta)],       # matrice de rotation (2x2)
                [sin(teta), cos(teta)]])

# Rotation des points
points_rotation = dot(R_teta, points)     # (2x2)x(2x1000): on trouve une matrice (2x1000)

# Bruit Gaussien 2D
bruit = randn(2,n) * 0.5                 # matrice (2x1000)

X = (points_rotation + bruit).T          # jeu de données final : une addition de matrice (2x1000) nous donne une matrice (2x1000)
                                         # la transposée parce qu'on veut une matrice (1000x2) : 1000 données  et deux variables


# Partie B : Mahalanobis et PAC
# Moyenne et la matrice de covariance
mu = mean(X, axis=0)

# soit "ecart", la différence entre nos données et la moyenne
ecart = (X-mu)
ecart_t = transpose(ecart)

sigma = dot(ecart, ecart_t) / (n-1)
sigma_inv = inv(sigma)

# Distance de Mahalanobis (formule déjà donné)
D = sqrt(sum(dot(ecart, sigma_inv) * ecart_t, axis=1))

# PAC
```
